```python
import pandas as pd
import numpy as np
!pip install textblob
from textblob import TextBlob
import re
from vaderSentiment.vaderSentiment import SentimentIntensityAnalyzer
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, classification_report
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis
```

```python
df1=pd.read_csv('C:\\Users\\Payal P\\Assingment_CS888\\Combined_News_DJIA.csv')
df1.head()
```

```python
df2= pd.read_csv('C:\\Users\\Payal P\\Assingment_CS888\\upload_DJIA_table.csv')
df2.head()
```

```python
df1.shape, df2.shape
```

```python
merge= df1.merge(df2, how='inner', on='Date')
merge
```

```python
headline= []

for row  in range (0, len(merge.index)):
    headline.append(" ".join(str(x) for x in merge.iloc[row, 2:27]))
```

```python
headline[0]
```

```python
clean_headline = []

for i in range (0, len(headline)):
    clean_headline.append(re.sub("b[(')]", '', headline[i])) # remove b'
    clean_headline[i] = re.sub('b[(")]', '', clean_headline[i]) # remove b"
    clean_headline[i] = re.sub("\'", '', clean_headline[i]) # remove \'
```

```python
clean_headline[0]
```

```python
merge['Combined_News'] = clean_headline

merge
```

```python
# get subjectivity:

def get_subjectivity(text):
    return TextBlob(text).sentiment.subjectivity

# get polarity:

def get_polarity(text):
    return TextBlob(text).sentiment.polarity
```

```python
merge['Subjectivity'] = merge['Combined_News'].apply(get_subjectivity)

merge['Polarity'] = merge['Combined_News'].apply(get_polarity)
```

```python
merge
```

```python
# create function to get the sentiment score:

def getSIA(text):
    sia = SentimentIntensityAnalyzer()
    sentiment= sia.polarity_scores(text)
    return sentiment
```

```python
# get Sentiment score for each day:
compound=[]
neg=[]
pos=[]
neu=[]
SIA=0

for i in range (0, len(merge['Combined_News'])):
    SIA= getSIA(merge['Combined_News'][i])
    compound.append(SIA['compound'])
    neg.append(SIA['neg'])
    pos.append(SIA['pos'])
    neu.append(SIA['neu'])
```

```python
merge['compound']= compound
merge['neg']= neg
merge['pos']=pos
merge['neu']=neu

merge
```

```python
df= merge[['Label','Open', 'High', 'Low','Volume','Subjectivity','Polarity','compound','neg','pos','neu']]
df
```

```python
#create the feature data set:
X=df

X=np.array(X.drop(['Label'],1))
```

```python
#create the target data set:

y= np.array(df['Label'])
```

```python
# Split the data:

x_train, x_test, y_train, y_test= train_test_split(X,y, test_size=0.2, random_state= 0)
```

```python
#create and train the model:

model= LinearDiscriminantAnalysis().fit(x_train, y_train)

# Model predicton:

predict= model.predict(x_test)

predict
```

```python
print(classification_report(y_test, predict))
```
